```python
batting_first=ipl[ipl["result_margin"]>10]
batting_first.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first["result_margin"])
plt.title("winning margine of batting first")
plt.xlabel("margine of runs")
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(5,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()), batting_first['winner'].value_counts()[0:3], color=["b", "g", "grey"])
plt.title('won when batted first')
plt.xlabel("names")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(batting_first['winner'].value_counts(),labels=list(batting_first['winner'].value_counts().keys()), autopct="%0.1f%%")
plt.show()
```
